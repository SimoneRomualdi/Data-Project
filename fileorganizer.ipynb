```python
import os
import shutil
import csv

# Definisco il percorso della cartella principale dove si trovano i file
directory = 'files'

# Definisco quali estensioni appartengono a immagini, audio e documenti
image_extensions = ['.png', '.jpg', '.jpeg']
audio_extensions = ['.mp3', '.wav']
doc_extensions = ['.txt', '.odt']

# Associo ogni tipo di file alla sua sottocartella dedicata
folders = {
    'image': 'images',
    'audio': 'audio',
    'doc': 'docs'
}

# Definisco il percorso del file recap.csv che userò per tracciare i file
recap_file = os.path.join(directory, 'recap.csv')

# Creo una funzione per determinare il tipo di file a partire dalla sua estensione
def get_file_type(extension):
    if extension in image_extensions:
        return 'image'
    elif extension in audio_extensions:
        return 'audio'
    elif extension in doc_extensions:
        return 'doc'
    else:
        return None

# Mi assicuro che le sottocartelle esistano già, altrimenti le creo
for folder in folders.values():
    folder_path = os.path.join(directory, folder)
    os.makedirs(folder_path, exist_ok=True)

# Leggo tutti i file presenti nella cartella principale (escludendo le cartelle)
files = sorted([f for f in os.listdir(directory) if os.path.isfile(os.path.join(directory, f))])

# Se il file recap.csv non esiste, lo creo e scrivo l'intestazione delle colonne
if not os.path.exists(recap_file):
    with open(recap_file, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['name', 'type', 'size'])

# Ora elaboro ogni file trovato
for file in files:
    file_path = os.path.join(directory, file)
    file_name, file_ext = os.path.splitext(file)

    # Determino il tipo di file
    file_type = get_file_type(file_ext.lower())

    # Se riesco a riconoscere il tipo di file, procedo
    if file_type:
        # Calcolo la dimensione del file in byte
        size = os.path.getsize(file_path)

        # Stampo a video il nome, il tipo e la dimensione del file
        print(f"{file_name} type:{file_type} size:{size}B")

        # Costruisco il percorso della destinazione dove spostare il file
        dest_folder = os.path.join(directory, folders[file_type])
        dest_path = os.path.join(dest_folder, file)

        # Sposto fisicamente il file nella sua sottocartella
        shutil.move(file_path, dest_path)

        # Apro il file recap.csv in modalità append e aggiungo una nuova riga con i dettagli
        with open(recap_file, 'a', newline='') as csvfile:
            writer = csv.writer(csvfile)
            writer.writerow([file_name, file_type, f"{size}B"])

# Fine STEP 1
```

```python
import os
import numpy as np
from PIL import Image
from tabulate import tabulate

# Definisco la cartella dove si trovano le immagini
images_folder = 'files/images'

# Creo una lista per memorizzare le righe della tabella
table_data = []

# Itero su tutti i file presenti nella cartella delle immagini
for filename in os.listdir(images_folder):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
        
        # Costruisco il path completo dell'immagine
        image_path = os.path.join(images_folder, filename)

        # Apro l'immagine con PIL
        image = Image.open(image_path)

        # Converto l'immagine in un array NumPy
        img_array = np.array(image)

        # Ricavo nome base (senza estensione)
        name = os.path.splitext(filename)[0]

        # Ottengo altezza e larghezza
        height, width = img_array.shape[0], img_array.shape[1]

        # Inizializzo i valori per la riga della tabella
        grayscale_mean = 0.0
        r_mean = g_mean = b_mean = alpha_mean = 0.0

        # Controllo il numero di dimensioni dell'array
        if img_array.ndim == 2:
            # Immagine in scala di grigi
            grayscale_mean = round(np.mean(img_array), 2)
        elif img_array.ndim == 3:
            channels = img_array.shape[2]
            if channels == 3:
                # Immagine RGB
                r_mean = round(np.mean(img_array[:, :, 0]), 2)
                g_mean = round(np.mean(img_array[:, :, 1]), 2)
                b_mean = round(np.mean(img_array[:, :, 2]), 2)
            elif channels == 4:
                # Immagine RGBA
                r_mean = round(np.mean(img_array[:, :, 0]), 2)
                g_mean = round(np.mean(img_array[:, :, 1]), 2)
                b_mean = round(np.mean(img_array[:, :, 2]), 2)
                alpha_mean = round(np.mean(img_array[:, :, 3]), 2)

        # Costruisco la riga da aggiungere alla tabella
        table_data.append([
            name,
            height,
            width,
            grayscale_mean,
            r_mean,
            g_mean,
            b_mean,
            alpha_mean
        ])

# Definisco le intestazioni della tabella
headers = ["name", "height", "width", "grayscale", "R", "G", "B", "ALPHA"]

# Stampo la tabella in formato elegante con tabulate
print(tabulate(table_data, headers=headers, tablefmt="fancy_grid"))
```
